GoGreenRide, a New York City-based car service, is a transportation startup that generates estimated pickup times and trip lengths based on historical trip data provided by Tom Tom's Map Toolkit API. A major target variable that is a key service performance indicator is the trip duration delta and lateness variable, which is the variation of the quoted time and the actual time. The trip duration delta is important because they directly impact ride revenue, while the lateness variable affects our reliability in the eyes of the customer. In this case, the benefit of the predictor is accurate pricing and an improved reliability reputation amongst customers.

# Supervised Learning Linear Regression Model

Every minute that is optimized represents a 5% boost to revenue. As such, the closer that our predictor can optimize the trip length, the higher the booked revenue.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from sklearn import linear_model
```

```python
data = pd.read_csv('project_data_v1.csv', error_bad_lines=False, dtype={'duration_delta': np.float32})
```

```python
expected_revenue_per_min = data['fare']/data['expected_duration'] 
actual_revenue_per_min = data['fare']/data['actual_duration']
x_val = expected_revenue_per_min.index

#plt.xlim(0, 10) 
plt.figure(figsize=(20,8))
plt.plot(x_val[12000:12100], expected_revenue_per_min[12000:12100], label="Expected Revenue")
plt.plot(x_val[12000:12100],actual_revenue_per_min[12000:12100], label="Actual Revenue")
plt.legend()
plt.show()
```

# Predicting Likelihood of being late based on Ride Type, Time of Day and Day of Week

Features Variables: <br>
   <ul>
   <li>TOD = Time of Day</li>
    <li>DOW = Day of Week</li>
    <li>Ride Numerical= Ride Type: 0 indicates Ride Now, 1 indicates Ride Later</li>
    </ul>

```python
data['islate'] = (data.minutes_late_to_pu > 0).astype(int)
```

```python
# Get our X Variables
X_Variables = ['ride_numerical','tod','dow']
X = data[X_Variables]
```

```python
X = X.values
```

```python
y = data['islate'].values
```

```python
clf = linear_model.LogisticRegression()
```

```python
model = clf.fit(X,y)
```

```python
model.score(X,y)
```
